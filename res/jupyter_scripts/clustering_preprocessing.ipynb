```python
import pandas as pd
```

```python
df = pd.read_csv('../clustering/migros-barcode-category.csv')
```

```python
serie = df['slug'].astype(str)
arr = []
for elem in serie:
    splitted = elem.split('/')
    arr.append(splitted[2:4])
    
clusters = pd.DataFrame(arr).dropna().drop_duplicates()
clusters = clusters[clusters[1].notnull()]
clusters = clusters.reset_index().drop('index',axis=1)


clusters.to_csv('clusters.csv')
```

```python
clusters['cluster_id'] = clusters.index
clusters
```

```python
df_clean = pd.concat([df,pd.DataFrame(arr,df.index)],axis=1)[['name','eans',0,1]].dropna()
```

```python
adri_file = pd.merge(df_clean, clusters,on=[0,1])
len(adri_file)
```

```python
adri_file.to_csv('../clustering/product_to_cluster.csv')
```
